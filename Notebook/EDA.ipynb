```python
#improting necessary libaries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#reading the csv file
data=pd.read_csv("../Data/Data_COPD.csv")
print(data.head())
```

```python
print(data.dtypes)
```

## Univariate Analysis

```python
data_cols=data.select_dtypes(include=['int64','float64']).columns
# Set up the figure and axes
fig, axes = plt.subplots(len(data_cols)//3 + 1, 3, figsize=(15, 15))
axes = axes.flatten()

# Plot each column as a histogram
for i, col in enumerate(data_cols):
    sns.histplot(data[col], kde=False, ax=axes[i])
    axes[i].set_title(f'Histogram of {col}')
plt.tight_layout()
plt.show()
```

```python
#seprarting categorical
data_cat=data.select_dtypes(include=['object','category','bool']).columns
print(data_cat)
```

```python
# Univariate analysis for the categorical features
# gender distribution
sns.countplot(x = 'Gender', data = data, palette='viridis')
plt.title("Gender Distribution")
```

```python
sns.countplot(x = 'Smoking_Status', data = data, palette='viridis')
plt.title("Smoking_Status")
```

```python
plt.figure(figsize=(10,10))
sns.countplot(x = 'Location' , data = data, palette='viridis')
plt.title('Location' ) 
```

## Bivariate Analysis

```python
# pair plot
sns.pairplot(data, hue='COPD_Diagnosis', diag_kind='kde', palette='Set1')
plt.title("Pair Plot")
plt.show()
```

```python
sns.boxplot(x = 'COPD_Diagnosis', y = 'Age', data = data, palette='coolwarm')
plt.title("Age vs COPD Diagnosis")
plt.show()
```

```python
plt.figure(figsize=(12,12))
sns.boxplot(data)
```

```python
 
sns.boxplot(x = 'COPD_Diagnosis', y = 'BMI', data = data, palette="coolwarm")
plt.title("BMI vs COPD_Diagnosis")
plt.show()
```

```python
sns.boxplot(x = 'Smoking_Status', hue='COPD_Diagnosis', data = data, palette='Set2')
plt.title("Smoking Status vs COPD_Diagnosis")
plt.show()
```

```python
sns.boxplot(x = 'Gender', y = 'BMI', data = data, palette="coolwarm")
plt.title("GENDER vs COPD_Diagnosis")
plt.show()
```
